```python
maoshlar = {"Ali": 800, "Vali": 1200, "Olim": 950, "Hasan": 1500, "Husan": 1100}
print("Ma'lumotlar muvaffaqiyatli kiritildi!")
```

```python
name, salary = max(maoshlar.items(), key=lambda item: item[1])
print(f"Eng ko'p maosh oladigan xodim: {name}, maoshi: {salary}")
```
